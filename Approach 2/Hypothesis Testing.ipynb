```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
# Load the data
df_professor = pd.read_csv('lstm_pred_final_3.csv')
df_new_approach = pd.read_csv('approach2_with_preds.csv')
df_new_approach['Profit'] = df_new_approach['Profit'].astype('Float64').round(2)
```

```python
# Drop any NaN profits, as they may represent trades not taken or incomplete data
trades_professor = df_professor[(df_professor['model_RMSE'] < 0.05)]
trades_new_approach = df_new_approach[(df_new_approach['model_RMSE'] < 0.05)]

profits_professor = trades_professor['Profit']
profits_new_approach = trades_new_approach['Profit']
```

```python
# Calculate variance for each group
variance_professor = np.var(profits_professor, ddof=1)  # Using ddof=1 for sample variance
variance_new_approach = np.var(profits_new_approach, ddof=1)  # Using ddof=1 for sample variance

# Print the variances
print("Variance of Group using Professor's Approach:", variance_professor)
print("Variance of Group using New Approach:", variance_new_approach)
```

```python
# Perform welch t-test (independent)
t_stat, p_value = stats.ttest_ind(profits_professor, profits_new_approach, equal_var=False)

# Set significance level
alpha = 0.05

# Print results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference after introducing the LSTM model.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference after introducing the LSTM model.")
```

```python
min_profit = round(profits_new_approach.min(), 2)
max_profit = round(profits_new_approach.max(), 2)

print('Minimum Profit:', min_profit, '%')
print('Maximum Profit:', max_profit, '%')
```
